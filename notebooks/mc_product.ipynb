```python
dfa_prism = stormpy.parse_prism_program(dfa)  # type: ignore
options = stormpy.BuilderOptions()  # type: ignore
options.set_build_all_labels()
options.set_build_choice_labels()
options.set_build_state_valuations()
dfas = stormpy.build_sparse_model_with_options(dfa_prism, options)  # type: ignore
dfa: Model = stormvogel.mapping.stormpy_to_stormvogel(dfas) # type: ignore
show.show(dfa)
```
